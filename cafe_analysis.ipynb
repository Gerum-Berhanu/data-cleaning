### Payment method through time

```python
df["Month"] = (
    pd.to_datetime(df["Transaction Date"], errors="coerce")
    .dt.to_period("M")
    .dt.strftime("%b")
)
```

```python
payment_method_counts_by_month = (
    df.groupby(["Month", "Payment Method"], dropna=False)
      .size()
      .unstack(fill_value=0)
)

display(payment_method_counts_by_month)
```

```python
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
payment_order = ["Cash", "Credit Card", "Digital Wallet", np.nan]

plot_df = payment_method_counts_by_month.reindex(index=month_order + [np.nan], columns=payment_order, fill_value=0)

x = np.arange(len(plot_df.index))
bar_width = 0.2

plt.figure(figsize=(14, 6))

for i, method in enumerate(plot_df.columns):
    plt.bar(x + i * bar_width, plot_df[method].values, width=bar_width, label=str(method))

plt.xticks(x + bar_width * (len(plot_df.columns) - 1) / 2,
           [m if pd.notna(m) else "NaN" for m in plot_df.index])

plt.xlabel("Month")
plt.ylabel("Count")
plt.title("Payment Method Counts by Month")
plt.legend(title="Payment Method")
plt.tight_layout()
plt.show()
```
