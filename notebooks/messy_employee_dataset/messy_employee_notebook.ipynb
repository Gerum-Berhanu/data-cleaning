### `phone`

```python
df["phone"].describe()
```

```python
df["phone"] = df["phone"].abs().astype(str)
```

```python
def fill_phone_with_zeros(phone: str):
    """Prepend zeros so that the phone number is a length of 10.
    
    For a given phone number of length 10 or greater, the number itself is returned back.
    """
    length = len(phone)
    zeros_count = 10 - length
    zeros_str = "0" * zeros_count
    return zeros_str + phone
```

```python
df["phone"] = df["phone"].apply(lambda x: fill_phone_with_zeros(x))
df["phone"] = df["phone"].apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}")
```

```python
invalid_phone_mask = ~df["phone"].str.match(r"\d{3}-\d{3}-\d{4}")

print("Count of phone numbers that don't abide the general format:", df.loc[invalid_phone_mask, "phone"].shape[0])
```

## Interpolation with Basic EDA

```python
def find_max_val_indices(table: pd.DataFrame):
    max_results = []

    for col in table.columns:
        max_val = table[col].max()
        down_max = table[col][table[col].lt(max_val)].max()

        # isolation_ratio = max_val / down_max if pd.notna(down_max) and down_max != 0 else 0
        isolation_diff = max_val - down_max if pd.notna(down_max) else -1
        
        tied_indices = table[table[col] == max_val].index.tolist()

        max_results.append({
            "Column": col,
            "Count": int(max_val),
            "Indices": tied_indices,
            "Isolation Diff": int(isolation_diff)
        })
    
    return max_results
```

### `age` vs `full_name`

```python
table_name_age = df.pivot_table(
    index="full_name",
    columns="age",
    aggfunc="size",
    fill_value=0
)

display(table_name_age)
```

```python
maxs_name_age = find_max_val_indices(table_name_age)

pprint(maxs_name_age, width=100, indent=2)
```
