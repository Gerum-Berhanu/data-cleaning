### Unstacking `Contract`

```python
def categorize_contract(contract_str):
    if pd.isna(contract_str):
        return "Unknown"
    elif "On Loan" in contract_str:
        return "On Loan"
    elif "Free" in contract_str:
        return "Free"
    else:
        return "Permanent"
    
df_info["Contract Type"] = df_info["Contract"].apply(categorize_contract)

display(df_info[["Contract", "Contract Type", "Loan Date End"]].head())
```

```python
def get_contract_year(contract_str, year_idx):
    # year_idx 0 for start, 1 for end
    if pd.isna(contract_str):
        return None
    
    if "~" in str(contract_str):
        year = str(contract_str).split("~")[year_idx].strip()
        return year
    return None

df_info["Contract Start Year"] = df_info["Contract"].apply(get_contract_year, year_idx=0)
df_info["Contract End Year"] = df_info["Contract"].apply(get_contract_year, year_idx=1)

display(df_info[["Contract", "Contract Start Year", "Contract End Year"]].head())
```

```python
df_info.drop(columns=["Contract"], inplace=True)
```

```python
first_order = df_info.loc[:, :"Joined"].columns.tolist()
second_order = ["Contract Type", "Contract Start Year", "Contract End Year"]
third_order = df_info.loc[:, "Loan Date End":"Release Clause"].columns.tolist()

df_info = df_info[first_order + second_order + third_order]

display(df_info.head(2))
```

### Dropping `Contract Start Year`

```python
joined_contract = df_info[["Joined", "Contract Start Year"]][df_info["Contract Start Year"].notna()]

extracted_year = joined_contract["Joined"].str.extract(r"(?<=,\s)(\d{4})")[0]
condition = (extracted_year == joined_contract["Contract Start Year"])

print("Count of unmatching year of `Joined` and `Contract Start Year` column:", (~condition).sum())
```

This means, `Joined` and `Contract Start Year` are 100% identical across the entire dataset. Due to this, it is redundant to keep both. So, we should drop the `Contract Start Year` column. Why not `Joined`? Because it contains more precise info (the month and the date).

```python
df_info.drop(columns="Contract Start Year", inplace=True)
```

#### `Height` to `float`

```python
height_format_mask = df_info["Height"].str.match(r"\d+cm")
outlier_height = df_info[~height_format_mask]["Height"]

print("Count of values that don't follow the \\d+cm format:", len(outlier_height))

display(outlier_height.head())
```

```python
print("Count of outliers matching the \\d+\'\\d+\" format:", df_info["Height"].str.match(r"\d+\'\d+\"").sum())
```

Because the above print value 40 is equal to the length of the total outlier formats, we can say all the outliers follow the same format which is `\d+'\d+"` (in feet and inches rather than centimeters).

```python
def to_cm_float(height_str: str):
    if pd.isna(height_str):
        return pd.NA
    
    height_str = height_str.strip().lower()
    
    if "'" in height_str:
        foot, inch = height_str.split("'")

        foot = float(foot)
        inch = float(inch.replace('"', ''))

        total_inch = (foot * 12) + inch

        return round(total_inch * 2.54, 2)
    
    elif "cm" in height_str:
        return round(float(height_str.replace("cm", "")), 2)
    
    else:
        return float(height_str)

df_info["Height"] = df_info["Height"].apply(to_cm_float)

df_info["Height"].head()
```

#### `Weight` to `float`

```python
weight_format_mask = df_info["Weight"].str.match(r"\d+kg")
outlier_weight = df_info[~weight_format_mask]["Weight"]

print("Count of values that don't follow the \\d+kg format:", len(outlier_weight))

display(outlier_weight.head())
```

```python
print("Count of outliers matching the \\d+lbs format:", len(outlier_weight.str.match(r"\d+lbs")))
```

All the weight outliers follow the `\d+lbs` format (in pound rather than kilograms).

```python
def to_kg_float(weight_str: str):
    if pd.isna(weight_str):
        return pd.NA
    
    weight_str = weight_str.strip()

    if "lbs" in weight_str:
        pound = float(weight_str.replace("lbs", ""))
        return round(pound * 0.454, 2)
    
    elif "kg" in weight_str:
        return round(float(weight_str.replace("kg", "")), 2)
    
    else:
        return float(weight_str)
    
df_info["Weight"] = df_info["Weight"].apply(to_kg_float)

df_info["Weight"].head()
```

#### `Value`, `Wage`, `Release Clause` to `float`

```python
value_format_mask = df_info["Value"].str.match(r"€\d+(?:\.\d+)?[KM]")
wage_format_mask = df_info["Wage"].str.match(r"€\d+(?:\.\d+)?[KM]")
clause_format_mask = df_info["Release Clause"].str.match(r"€\d+(?:\.\d+)?[KM]")

outlier_value = df_info[~value_format_mask]["Value"]
outlier_wage = df_info[~wage_format_mask]["Wage"]
outlier_clause = df_info[~clause_format_mask]["Release Clause"]

print("Count of values that don't follow the €\\d+(?:\\.\\d+)?[KM] format:")
print("\tValue:", len(outlier_value))
print("\tWage:", len(outlier_wage))
print("\tRelease Clause:", len(outlier_clause))

display(outlier_value.head(2))
display(outlier_wage.head(2))
display(outlier_clause.head(2))

display(outlier_value.tail(2))
display(outlier_wage.tail(2))
display(outlier_clause.tail(2))
```

```python
print(outlier_value.unique())
print(outlier_wage.unique())
print(outlier_clause.unique())
```

```python
def to_million_float(value_str: str):
    if pd.isna(value_str):
        return pd.NA
    
    value_str = value_str.strip().replace("€", "")

    if "M" in value_str:
        value_float = float(value_str.replace("M", "")) * 10e6
        return value_float
    
    elif "K" in value_str:
        value_float = float(value_str.replace("K", "")) * 10e3
        return value_float

    else:
        return float(value_str)
    
df_info["Value"] = df_info["Value"].apply(to_million_float)
df_info["Wage"] = df_info["Wage"].apply(to_million_float)
df_info["Release Clause"] = df_info["Release Clause"].apply(to_million_float)

display(df_info["Value"].head())
display(df_info["Wage"].head())
display(df_info["Release Clause"].head())
```
