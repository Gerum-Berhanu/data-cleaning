## Whitespace and Duplicate Cleaning

```python
# removing trailing whitespaces and non-ASCII characters
df.columns = df.columns.str.strip().str.replace(r"[^\x00-\x7F]", "", regex=True)

# inserting an underscore between lowercase and uppercase letters (playerUrl -> player_Url)
df.columns = df.columns.str.replace(r'([a-z])([A-Z])', r'\1 \2', regex=True)

# ensuring the first letter of every column name is in uppercase
df.columns = [col[0].upper() + col[1:] for col in df.columns]

print(df.columns)
```

```python
print("Duplicates:",df.duplicated().sum())
```

### Unstacking Contract

```python
def categorize_contract(contract_str):
    if pd.isna(contract_str):
        return "Unknown"
    elif "On Loan" in contract_str:
        return "On Loan"
    elif "Free" in contract_str:
        return "Free"
    else:
        return "Permanent"
    
df_info["Contract Type"] = df_info["Contract"].apply(categorize_contract)

display(df_info[["Contract", "Contract Type", "Loan Date End"]])
```

```python
def get_contract_year(contract_str, year_idx):
    # year_idx 0 for start, 1 for end
    if pd.isna(contract_str):
        return None
    
    if "~" in str(contract_str):
        year = str(contract_str).split("~")[year_idx].strip()
        return year
    return None

df_info["Contract Start Year"] = df_info["Contract"].apply(get_contract_year, year_idx=0)
df_info["Contract End Year"] = df_info["Contract"].apply(get_contract_year, year_idx=1)

display(df_info[["Contract", "Contract Start Year", "Contract End Year"]].head())
```

```python
df_info.drop(columns=["Contract"], inplace=True)
```

```python
first_order = df_info.loc[:, :"Joined"].columns.tolist()
second_order = ["Contract Type", "Contract Start Year", "Contract End Year"]
third_order = df_info.loc[:, "Loan Date End":"Release Clause"].columns.tolist()

df_info = df_info[first_order + second_order + third_order]
```
